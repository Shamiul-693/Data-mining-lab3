```python
import pandas as pd

# Load the data into a pandas dataframe
df = pd.read_csv('lenses.data', header=None, delim_whitespace=True, names=['age', 'prescription', 'astigmatism', 'tear_rate', 'lenses'])

```

```python
# Print the first 5 rows of the dataframe to verify that it loaded correctly
print(df.head())
```

#dicision tree

```python
# Import the necessary libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report



# Convert categorical variables to numerical using one-hot encoding
df = pd.get_dummies(df, columns=['age', 'prescription', 'astigmatism', 'tear_rate'])


```

```python
# Separate features and target variable
X = df.drop('lenses', axis=1)
y = df['lenses']

```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
# Create a Decision Tree classifier
clf = DecisionTreeClassifier()

```

```python
# Train the classifier on the training set
clf.fit(X_train, y_train)

# Test the classifier on the testing set
y_pred = clf.predict(X_test)

# Print the classification report
print(classification_report(y_test, y_pred))
```

```python
# Test the classifier on the testing set
y_pred = clf.predict(X_test)

```

```python
# Print the classification report
print(classification_report(y_test, y_pred))
```

#neural network

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import numpy as np

# Load the dataset into a numpy array
data = np.loadtxt('lenses.data', dtype=int)


```

```python
# Split the data into features (X) and target variable (y)
X = data[:, :-1]
y = data[:, -1]

```

```python
# Convert target variable to one-hot encoding
num_classes = 4  # There are four classes in the target variable
y = keras.utils.to_categorical(y, num_classes=num_classes)

```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
# Define the neural network architecture
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(32, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))  # Output layer has four neurons to match the number of classes

```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

```

```python
# Train the model on the training set
model.fit(X_train, y_train, epochs=80, batch_size=32)

```

```python
# Test the model on the testing set
y_pred_prob = model.predict(X_test)
y_predd = np.argmax(y_pred_prob, axis=1)

```

```python
# Print the classification report
print(classification_report(np.argmax(y_test, axis=1), y_predd))
```
